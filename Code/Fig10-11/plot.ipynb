```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xgboost = [0.77,0.83, 0.85, 0.89]
DT = [0.68,0.76, 0.77, 0.84]
Bagging = [0.73,0.80, 0.73, 0.87]
KNN = [0.48,0.52, 0.50, 0.57]
index = [5, 10, 15, 20]

 

df = pd.DataFrame({"XGboost": xgboost, 'Bagging': Bagging,'DT': DT, 'KNN': KNN}, index=index)

 

ax = df.plot.bar(rot=0)
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1),loc='upper left', fontsize=14)

 

ax.set_xlabel("Window size (To)",  fontsize=16)
ax.set_ylabel("Accuarcy",  fontsize=16)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.title('Accuracy scores with different window sizes',fontsize=18)
# ax.set_title("Time Window = 10 s", color= 'black')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
xgboost = [0.83,0.86, 0.88, 0.88]
DT = [0.81,0.83, 0.85, 0.86]
Bagging = [0.81,0.84, 0.86, 0.86]
KNN = [0.55,0.55, 0.56, 0.54]

 

index = [5, 10, 15, 20]

 

df = pd.DataFrame({"XGboost": xgboost, 'Bagging': Bagging,'DT': DT, 'KNN': KNN}, index=index)

 

ax = df.plot.bar(rot=0)
ax.grid()
ax.legend(bbox_to_anchor=(1.05, 1),loc='upper left', fontsize=14)

 

ax.set_xlabel("Window size (To)",fontsize=16)
ax.set_ylabel("AUC",fontsize=16)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.title('Area under the ROC curve with different window sizes',fontsize=18)
# ax.set_title("Time Window = 10 s", color= 'black')
```
